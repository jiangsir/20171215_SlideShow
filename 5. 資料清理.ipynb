# 資料清理

## 用 pandas 讀入 excel 進行處理。

* 資料轉換：民國年->西元年
* 別名統一化 : 台大，臺大，台灣大學，國立台灣大學
* 舊名換新名：高雄縣->高雄市, 台北縣->新北市
* 資料型態異常：０９２０１２３４５６ -> 0920123456
* 不規範的日期格式，民國西元混雜
* 遺失值的處理


```python
import pandas as pd
basepath = 'datas/'

df = pd.read_excel(basepath + '清理資料.xlsx')
df
```

```python
# 丟掉有 NaN 的 row
df.dropna()
```

```python
# NaN 的值統一設定成某一個特定資料。

df.證書日期 = df[['證書日期']].fillna('20171215')
df.head(10)
```

```python
# 姓名馬賽克

names = []
for name in df.姓名:
    names.append(name[0]+"O"+name[2])

df.姓名 = names
df.head()
```

## 聽說資料分析有 80% 的時間會花在資料清理上。

```python
import datetime

pattern = '[0-9]{2,3}/[0-9]+/[0-9]+' # 篩選出  103/06/02
pattern1 = '[0-9]{2,3}\.[0-9]+\.[0-9]+'

df['pattern'] = df.證書日期.str.contains(pattern)

dates = []
for index, row in df[df['pattern']==True].iterrows():
    year = int(row['證書日期'].split('/')[0])+1911
    m = int(row['證書日期'].split('/')[1])
    d = int(row['證書日期'].split('/')[2])
    # print(row['證書日期'], int(row['證書日期'].split('/')[0])+1911)
    date = datetime.datetime.strptime(str(year)+'/'+str(m)+'/'+str(d), '%Y/%m/%d').date()
    dates.append(date)

df.loc[df['pattern']==True, '證書日期'] = dates

df
```
